```python
import random

# Proseguiamo scrivendo la logica del codice, definiamo le funzioni


scelta_valide = ["sasso", "carta", "forbice"]

while True: # Il ciclo si ripete finché non viene interrotto esplicitamente
    # La funzione input() accetta un solo argomento per il prompt.
    utente = input("Scegli tra sasso, carta, o forbice (o 'exit' per uscire): ").lower()

    # Se l'utente vuole uscire, gestiamo l'uscita prima di altre validazioni.
    if utente == "exit":
        print("È stato un piacere giocare con te, saluti!")
        break # Esci dal ciclo while
    # Verifichiamo la scelta dell'utente per le opzioni valide del gioco
    elif utente not in scelta_valide:
        print("Scelta non valida. Per favore, scegli tra sasso, carta o forbice.")
    else:
        # Il robot sceglie una delle opzioni valide.
        robot = random.choice(scelta_valide)

        print(f"La tua scelta: {utente}")
        print(f"La scelta del robot: {robot}")

        if utente == robot:
            print("È un pareggio!")
        elif (
            (utente == "sasso" and robot == "forbice") or
            (utente == "carta" and robot == "sasso") or
            (utente == "forbice" and robot == "carta")
        ):
            print("Hai vinto!")
            print("vuoi giocare ancora?")

        else:
            print("Hai perso!")
```
